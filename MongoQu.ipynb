```python
import pymongo
import pprint
```

```python
db = pymongo.MongoClient().production
```

```python
nodes = db.washdc.aggregate( [
                                { "$match": { "type": "node" } }, 
                                { "$group": { "_id": "$type", "count": { "$sum": 1 } } } 
                              ] )

for node in nodes:
    print(node)
```

```python
ways = db.washdc.aggregate( [
                                { "$match": { "type": "way" } }, 
                                { "$group": { "_id": "$type", "count": { "$sum": 1 } } } 
                              ] )

for way in ways:
    print(way)
```

```python
users = db.washdc.aggregate( [ 
                                { "$group": { "_id": "$created.user", "count": { "$sum": 1 } } },
                                { "$sort": { "count": -1 } }
                              ] )
```

```python
for user in users:
    print(user)
```

```python
len(db.washdc.distinct( "created.user" ))
```

```python
names = db.washdc.aggregate( [ { "$match": { "name": { "$exists": 1 } } },
                               { "$match": { "$and": [ { "type": "way" }, { "name_type": { "$ne": "None" } } ] } },
                               { "$group": { "_id": "$name_type", "count": { "$sum": 1 } } } ] )
```

```python
names = db.washdc.aggregate( [ { "$match": { "name": { "$exists": 1 } } },
                               { "$match": { "type": "way" } },
                               { "$group": { "_id": "$name_type", "count": { "$sum": 1 } } },
                               { "$sort": {"count": -1}},
                               { "$limit": 5 }] )
```

```python
for name in names:
    print(name)
```

```python
print(db.washdc.find_one( {"type": "way", "name_type": "St"}))
```

```python
amenities = db.washdc.aggregate([{"$match": {"amenity": {"$exists": 1}}},
                                 {"$group": {"_id": "$amenity", "count": {"$sum": 1}}},
                                 {"$sort": {"count": -1}},
                                 {"$limit": 5}])
```

```python
for amenity in amenities:
    print(amenity)
```

```python
db.washdc.find({"amenity": {"$exists": 1}}).count()
```

```python
wrong_name = db.washdc.find({"name": {"$regex": "S+$", "$options": "i"}})

wrong_n = []

for name in wrong_name:
    wrong_n.append(name)
```

```python
for each in wrong_n:
    print(each)
```
